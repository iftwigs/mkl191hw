### Задание 2

```python
URL = "http://gramota.ru/class/coach/punct/45_177"
r = requests.get(URL) 
bs=BeautifulSoup(r.text, "html5lib") 
  
soup = BeautifulSoup(" ".join([p.text for p in bs.find_all("p")]), "html5lib").get_text().replace("\xa0", " ")
#print(soup)
soup2 = BeautifulSoup(" ".join([div.text for div in bs.find_all("li")]), "html5lib").get_text().replace("\xa0", " ")
#print(soup2)
soup3 = soup + soup2
print(soup3[:1000])
```

```python
lowered = soup3.lower()
lowered[:1000]
```

```python
def remove_ns(text):
    return re.sub(r'\n', '', text)
clean = remove_ns(lowered)
clean[:1000]
```

```python
def delete_punc(text):
    return re.sub(r'[().\–_,\':«»;…?!]', ' ', text)
clean2 = delete_punc(clean)
clean2[:1000]
```

```python
tokens = re.split(' ', clean2)
tokens[:100]
```

```python
stemmer = SnowballStemmer('russian')
```

```python
stemmed = []
for w in tokens:
    if w in tokens:
        stemmed.append((w, stemmer.stem(w)))     
```

```python
stemmed[:100]
```

```python
stemmed_clean = []
for i in range(len(stemmed)):
    if stemmed[i]!=('', ''):
        stemmed_clean.append(stemmed[i])
```

```python
stemmed_clean[:100]
```

```python
URL = "https://zona.media/news/2019/09/11/zhukov-expertise"
r = requests.get(URL) 
bs=BeautifulSoup(r.text, "html5lib") 

text2 = BeautifulSoup(" ".join([p.text for p in bs.find_all("p")]), "html5lib").get_text().replace("\xa0", " ")
print(soup)
```

```python
tokens = sent_tokenize(text2, 'russian')
```

```python
tokens
```

```python
clean_tokens = []
def cleaner(sents):
    for i in range(len(sents)):
        clean_tokens.append(re.findall('\w+', (re.sub(r'[«»<>()"—.\–_,\':;…?!]', '', sents[i].replace('\u200b', ' '))).lower()))
cleaner(tokens)
clean_tokens[:100]
```

```python
pymorphed = []
for i in range(len(clean_tokens)):
    for j in range(len(clean_tokens[i])):
        pymorphed.append(morph.parse(clean_tokens[i][j]))
pymorphed[:10]       
```
